# Simple Transformer

```python
def micro_f1_score(true_bools,pred_bools):
  micro_f1_score = metrics.f1_score(true_bools,pred_bools,average='micro')
  return micro_f1_score

def macro_f1_score(true_bools,pred_bools):
  macro_f1_score = metrics.f1_score(true_bools,pred_bools,average='macro')
  return macro_f1_score
```

```python
model_args = MultiLabelClassificationArgs(
    train_batch_size = 16,
    eval_batch_size = 16,
    overwrite_output_dir=True,
    save_steps=-1,
    save_model_every_epoch=False,
    learning_rate = 1e-5,
    num_train_epochs = 20,
    weight_decay = 5e-10,
    # scheduler = "cosine_schedule_with_warmup",
    no_save = True,
    gradient_accumulation_steps=16
)

def train_and_eval(model_args, df): 

  model = MultiLabelClassificationModel(
      "xlmroberta",
      "xlm-roberta-large",
      num_labels=len(all_labels),
      args=model_args
  )

  model.train_model(df)

  return model
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
model = train_and_eval(model_args, df_train)
```

```python
def get_all_languages():
    return ["ru", "po", "ka", "it", "gr", "ge", "fr", "en", "es"]
def get_predictions(model, df):
  for language in get_all_languages():
    df_language = df[df['language'] == language]

    to_predict = list(df_language['text'].to_numpy())
    indices = list(df_language.index.values)
    predictions, _ = model.predict(to_predict)

    with open(f'./to_submit_{language}.txt', 'w') as f:
      for index, prediction in zip(indices, predictions):
        predicted_labels = ",".join([all_labels[i] for i in range(len(prediction)) if prediction[i]])

        f.write(f"{index[0]}\t{index[1]}\t{predicted_labels}\n")
```

```python
get_predictions(model, df_test)
```
